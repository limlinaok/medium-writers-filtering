```python
import pandas as pd
df = pd.read_csv("clean_writer.csv")
df.head()
```

```python
# Sort out writers who had subscribed to Medium at least 2 times
medium_writers = df[df["subscribed_times"] >= 2]
medium_writers.head()
```

```python
# Check how many participants meet the condition after 1st criteria: medium subscription
len(medium_writers)
```

```python
# Writers whose articles are within 6 months (180 days)
import pandas as pd
# Convert article dates to datetime
df ["last_article_date"] = pd.to_datetime(df["last_article_date"], errors = 'coerce')

# Get today's date
today = pd.Timestamp.today()

#Apply criteria one
medium_writers = df[df["subscribed_times"] >= 2]

#Filter writers who also met medium subscription criteria
medium_recent_writers = medium_writers[(today - medium_writers["last_article_date"]).dt.days <= 180]

# Check the result
print("Eligible writers: ", len(medium_recent_writers))
medium_recent_writers.head(10)

```

```python
# Filter out writers who have at least 150 words article and topic relevance
eligible_articles = medium_recent_writers[
    (medium_recent_writers["avg_article_length"] >= 150) &
    (medium_recent_writers["topic_tags"].str.contains("AI|ML|Data|Deep Learning", case=False, na=False))
]
print("Eligible writers: ", len(eligible_articles))
eligible_articles.head(10)
```

```python
# Writers must have at least 5 personal projects 
personalprojects_eligible = eligible_articles[eligible_articles["personal_projects_count"] >=5]
#Check result
print("personalprojects_eligible: ",len(personalprojects_eligible))
```

```python
# Only accept writers who has Stripe or Paypal as payment method
payment_eligible = personalprojects_eligible [personalprojects_eligible["payment_method"].str.contains("Stripe|Paypal",case=False, na=False)]
print("Payment eligible writers: ", len(payment_eligible))
payment_eligible.head()
```

```python
payment_eligible.to_csv("eligible_writers.csv", index=False)
```
